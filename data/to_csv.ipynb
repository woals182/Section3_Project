```python
import pandas as pd
import sqlite3

import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_csv('2017.csv',encoding='euc-kr')
df2 = pd.read_csv('2018.csv',encoding='euc-kr')
df3 = pd.read_csv('2018_2019.csv',encoding='euc-kr')
df4 = pd.read_csv('2019_2020.csv',encoding='euc-kr')
df5 = pd.read_csv('2020_2021.csv',encoding='euc-kr')
df6 = pd.read_csv('2021_2022.csv',encoding='euc-kr')
```

```python
df1_c = df1.copy()
df2_c = df2.copy()
df3_c = df3.copy()
df4_c = df4.copy()
df5_c = df5.copy()
df6_c = df6.copy()
```

```python
###결측치를 확인하고 결측치를 0으로 처리하기
df1_fillna = df1_c.fillna(0)
df2_fillna = df2_c.fillna(0)
df3_fillna = df3_c.fillna(0)
df4_fillna = df4_c.fillna(0)
df5_fillna = df5_c.fillna(0)
df6_fillna = df6_c.fillna(0)
```

```python
### 서울지역 
seoul1 = df1_fillna[df1_fillna['지점']==108]
seoul2 = df2_fillna[df2_fillna['지점']==108]
seoul3 = df3_fillna[df3_fillna['지점']==108]
seoul4 = df4_fillna[df4_fillna['지점']==108]
seoul5 = df5_fillna[df5_fillna['지점']==108]
seoul6 = df6_fillna[df6_fillna['지점']==108]
```

```python
### 대전지역
dae1 = df1_fillna[df1_fillna['지점']==133]
dae2 = df2_fillna[df2_fillna['지점']==133]
dae3 = df3_fillna[df3_fillna['지점']==133]
dae4 = df4_fillna[df4_fillna['지점']==133]
dae5 = df5_fillna[df5_fillna['지점']==133]
dae6 = df6_fillna[df6_fillna['지점']==133]
```

```python
### 청주지역
che1 = df1_fillna[df1_fillna['지점']==131]
che2 = df2_fillna[df2_fillna['지점']==131]
che3 = df3_fillna[df3_fillna['지점']==131]
che4 = df4_fillna[df4_fillna['지점']==131]
che5 = df5_fillna[df5_fillna['지점']==131]
che6 = df6_fillna[df6_fillna['지점']==131]
```

```python
Seoul = pd.concat([seoul1,seoul2,seoul3,seoul4,seoul5,seoul6])
daejeon = pd.concat([dae1,dae2,dae3,dae4,dae5,dae6])
cheongju = pd.concat([che1,che2,che3,che4,che5,che6])
```

```python
Seoul_d=Seoul[['지점명','일시','기온(°C)','강수량(mm)','풍속(m/s)','풍향(16방위)','습도(%)']]
dae_d=daejeon[['지점명','일시','기온(°C)','강수량(mm)','풍속(m/s)','풍향(16방위)','습도(%)']]
che_d=cheongju[['지점명','일시','기온(°C)','강수량(mm)','풍속(m/s)','풍향(16방위)','습도(%)']]
```

```python
Seoul_d=Seoul_d.reset_index()
dae_d=dae_d.reset_index()
che_d=che_d.reset_index()
```

```python
Seoul_dr = Seoul_d.drop(labels = 'index', axis=1)
dae_dr = dae_d.drop(labels = 'index', axis=1)
che_dr = che_d.drop(labels = 'index', axis=1)
```

```python
Seoul_dr['doy'] = None
Seoul_dr['hour'] = None
dae_dr['doy'] = None
dae_dr['hour'] = None
che_dr['doy'] = None
che_dr['hour'] = None
```

```python
Seoul_dr['datetime']= pd.to_datetime(Seoul_dr['일시'])
dae_dr['datetime']= pd.to_datetime(dae_dr['일시'])
che_dr['datetime']= pd.to_datetime(che_dr['일시'])
```

```python
for i in Seoul_dr.index:
    Seoul_dr['doy'][i]=pd.Period(Seoul_dr['datetime'][i],freq='D')
    Seoul_dr['doy'][i]=Seoul_dr['datetime'][i].day_of_year
    dae_dr['doy'][i]=pd.Period(dae_dr['datetime'][i],freq='D')
    dae_dr['doy'][i]=dae_dr['datetime'][i].day_of_year
    che_dr['doy'][i]=pd.Period(che_dr['datetime'][i],freq='D')
    che_dr['doy'][i]=che_dr['datetime'][i].day_of_year
```

```python
for i in Seoul_dr.index:
    Seoul_dr['hour'][i]=pd.Period(Seoul_dr['datetime'][i],freq='D')
    Seoul_dr['hour'][i]=Seoul_dr['datetime'][i].hour
    dae_dr['hour'][i]=pd.Period(dae_dr['datetime'][i],freq='D')
    dae_dr['hour'][i]=dae_dr['datetime'][i].hour
    che_dr['hour'][i]=pd.Period(che_dr['datetime'][i],freq='D')
    che_dr['hour'][i]=che_dr['datetime'][i].hour
```

```python
Seoul_df=Seoul_dr[['기온(°C)','강수량(mm)','풍속(m/s)','풍향(16방위)','습도(%)','doy','hour','datetime']]
dae_df=dae_dr[['기온(°C)','강수량(mm)','풍속(m/s)','풍향(16방위)','습도(%)','doy','hour','datetime']]
che_df=che_dr[['기온(°C)','강수량(mm)','풍속(m/s)','풍향(16방위)','습도(%)','doy','hour','datetime']]
```

```python
Seoul_df=Seoul_df.set_index('datetime',drop=True,append=False,inplace=False)
dae_df=dae_df.set_index('datetime',drop=True,append=False,inplace=False)
che_df=che_df.set_index('datetime',drop=True,append=False,inplace=False)
```
